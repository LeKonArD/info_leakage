```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("combo_results.tsv", sep="\t", index_col=0)
```

```python
pallette_p = []
for d in data.Publisher:
    
    if d == "Bastei LÃ¼bbe":
        pallette_p.append("blue")
    if d == "Pabel Moewig":
        pallette_p.append("black")
    if d == "Martin Kelter Verlag":
        pallette_p.append("green")
    if d == "CORA Verlag":
        pallette_p.append("red")
```

```python
sns.set(font_scale=1.5)
```

```python
fig, ax = plt.subplots(figsize=(16,9))

sns.swarmplot(x="Genre",y="Accuracy", hue="Publisher", data=data, palette=["blue","black","green","red"], alpha=0.5)
sns.boxplot(x="Genre",y="Accuracy", hue="Combination", data=data, palette=["purple","orange"], boxprops=dict(alpha=.5))

plt.savefig("super_plot.png", format="png", dpi=150, bbox_inches="tight")
```
